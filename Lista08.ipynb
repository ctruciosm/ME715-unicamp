```python
!pip install linearmodels
!pip install wooldridge
import wooldridge as woo
import numpy as np
import statsmodels.formula.api as smf
import scipy.stats as stats
import statsmodels.api as sm
import linearmodels.iv as iv
```

```python
fertil2 = woo.dataWoo('fertil2')
fertil2
```

a)

```python
modelo = smf.ols(formula = 'children ~ educ + age + np.square(age)', data = fertil2)
results = modelo.fit()
results.params
```

b)

```python
modelo = smf.ols(formula = 'age ~ frsthalf', data = fertil2)
results = modelo.fit()

modelo1 = smf.ols(formula = 'np.square(age) ~ frsthalf', data = fertil2)
results1 = modelo1.fit()
results1.pvalues

modelo1 = smf.ols(formula = 'educ ~ frsthalf + age + np.square(age)', data = fertil2)
results1 = modelo1.fit()
results1.pvalues

print(results.pvalues, results1.pvalues)
```

c)

```python
modelo = smf.ols(formula = 'educ ~ frsthalf + age + np.square(age)', data = fertil2)
results = modelo.fit()

fertil2['educfitted']=results.fittedvalues

modelo = smf.ols(formula = 'children ~ educfitted + age + np.square(age)', data = fertil2)
results = modelo.fit()
results.params
```

d)

```python
fertil2 = fertil2.dropna()

modelo_MQO = smf.ols(formula = 'children ~ educ + age + I(age**2) + electric + tv + bicycle', data = fertil2)
results_MQO = modelo_MQO .fit()

modelo1 = smf.ols(formula = 'educ ~ frsthalf + age + I(age**2) + electric + tv + bicycle', data = fertil2)
results1 = modelo1.fit()
results1.params
fertil2['children_fitted'] = results1.fittedvalues

modelo2 = smf.ols(formula = 'frsthalf ~ children_fitted + age + I(age**2) + electric + tv + bicycle', data = fertil2)
results2 = modelo2.fit()

print(results_MQO.params, results2.params)
```
